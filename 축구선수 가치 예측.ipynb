```python
def con_period(x):
    if x == 'Dec 31, 2018' :
        return '2019'
    elif x == 'Jun 30, 2020' :
        return '2020.5'
    elif x == 'Jun 30, 2019' :
        return '2019.5'
    elif x == 'May 31, 2020' :
        return '2020.416'
    elif x == 'May 31, 2019' :
        return '2019.416'
    elif x == 'Jan 31, 2019' :
        return '2019.0833'
    elif x == 'Jan 1, 2019' :
        return '2019'
    elif x == 'Jan 12, 2019' :
        return '2019.034'
    elif x == 'Dec 31, 2019' :
        return '2020'
    elif x == 'Jun 1, 2019' :
        return '2019.416'
    else :
        return x
```

```python
fifa_train.contract_until = fifa_train.contract_until.apply(con_period).astype('float64')
fifa_test.contract_until = fifa_test.contract_until.apply(con_period).astype('float64')
```

```python
fifa_train['contract'] = 2018
fifa_test['contract'] = 2018
```

```python
fifa_train.contract_until -= fifa_train.contract 
fifa_test.contract_until -= fifa_test.contract 
```

```python
fifa_train.drop('contract', axis = 1, inplace = True)
fifa_test.drop('contract', axis = 1, inplace = True)
```

```python
X = fifa_train.drop(columns=['id','name','value'])
y = fifa_train['value']

from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor() 
rf.fit(X,y)
```

```python
x = fifa_test.drop(columns=['id','name'])
y_hat = rf.predict(x)

result = pd.read_csv('C:/Users/jeonu/OneDrive/바탕 화면/데이콘 발표/FIFA_test.csv')
result['value'] = y_hat
```
